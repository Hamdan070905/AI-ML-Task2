```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('AI ML Task2.csv')
print(df)
df.head()
```

```python
print(df.head())

print(df.info())

print(df.describe(include='all'))

print(df.isnull().sum())
```

```python
print(df.describe())
print(df.median(numeric_only=True))
print(df.std(numeric_only=True))
print(df.mean(numeric_only=True))
print(df.select_dtypes(include='object').describe())
```

```python
df.hist(bins=30, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns
for col in numeric_cols:
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df[numeric_cols])
plt.show()
```

```python
for col in df.select_dtypes(include='object'):
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Distribution of {col}')
    plt.show()

sns.boxplot(x='Name', y='Age', data=df)
plt.show()

```

```python
fig = px.histogram(df, x='Age', color='Sex')
fig.show()

fig = px.scatter(df, x='Fare', y='Age', color='Survived')
fig.show()
```

```python
cleaned_df = df.drop_duplicates().dropna()
cleaned_df = df.drop_duplicates().fillna(df.mean(numeric_only=True))
print(cleaned_df)

```
